```python

import os
import pickle
import numpy as np
import torch
from torchvision.utils import save_image

def generate_image(G_PKL, generate_image_path):

  # Paths (replace if yours differ)
  W1_NPZ = '/content/invert_results/input1/projected_w.npz'
  W2_NPZ = '/content/invert_results/input2/projected_w.npz'
  BASE_OUT_DIR = '/content/drive/MyDrive/Filtered Fashion images/men_generated_outfit'

  # Create the base output directory if it doesn't exist
  os.makedirs(BASE_OUT_DIR, exist_ok=True)
  print(f"Base output directory created or exists: {BASE_OUT_DIR}") # Debug print

  # Create the specific output directory for this generated image
  OUT_DIR = os.path.join(BASE_OUT_DIR, generate_image_path)
  print(f"Attempting to create output directory: {OUT_DIR}") # Debug print
  os.makedirs(OUT_DIR, exist_ok=True)
  print(f"Output directory created or exists: {OUT_DIR}") # Debug print

  # 1) Load the trained generator (G_ema)
  with open(G_PKL, 'rb') as f:
      G = pickle.load(f)['G_ema'].cuda()

  # 2) Load projected latents
  w1 = torch.from_numpy(np.load(W1_NPZ)['w']).cuda()  # [1, 14, 512]
  w2 = torch.from_numpy(np.load(W2_NPZ)['w']).cuda()

  # 3) Define three different mixes:

  # 3A: Swap layers 4–13 (mid + fine) from w2 into w1
  w_mix_4to13 = w1.clone()
  w_mix_4to13[:, 4:14, :] = w2[:, 4:14, :]

  # 3B: Swap only layers 0–3 (coarse/shape) from w2 into w1
  w_mix_0to3 = w1.clone()
  w_mix_0to3[:, 0:4, :] = w2[:, 0:4, :]

  # 3C: Swap only layers 8–13 (purely fine texture) from w2 into w1
  w_mix_8to13 = w1.clone()
  w_mix_8to13[:, 8:14, :] = w2[:, 8:14, :]

  # 4) Synthesize each mixed latent
  with torch.no_grad():
      img_4to13 = G.synthesis(w_mix_4to13, noise_mode='const')
      img_0to3  = G.synthesis(w_mix_0to3,  noise_mode='const')
      img_8to13 = G.synthesis(w_mix_8to13, noise_mode='const')

  # 5) Normalize to [0,1] and save
  for name, img in [('mixed_4to13.png', img_4to13),
                    ('mixed_0to3.png', img_0to3),
                    ('mixed_8to13.png', img_8to13)]:
      img = (img + 1) / 2

      # out_path = os.path.join("/content/drive/MyDrive/Filtered Fashion images/men_generated_outfit", out_path)
      save_image(img, os.path.join(OUT_DIR, name))

  print("Saved three mixed outputs in:", OUT_DIR)
```
